#### Pipes

```python
# Categorical and Numerical features
cat_cols = ['Clear', 'Foggy', 'Mostly Cloudy', 'Overcast', 'Partly Cloudy']
num_cols = list(set(X_train.columns) - set(cat_cols))
```

```python
# Defining Pipelines
cat_pipe = Pipeline(steps = [('encoder', OrdinalEncoder()) #if the features were not encoded
                            ])

num_pipe = Pipeline(steps=[
                ('impute', KNNImputer(n_neighbors=5)), #if there are any missing values
                ('scaler', MinMaxScaler())
])

preprocessor = ColumnTransformer(transformers=[
                ('categorical',cat_pipe,cat_cols),
                ('numerical', num_pipe,num_cols)],
                 remainder='drop'
                )
```

## Model Training

#### GridSearchCV

```python
def model_loop(X, y, reg, grid_params):
    
    print('Currently fitting model:', reg)
    
    # Define KF
    kf = KFold(n_splits=3,shuffle=True, random_state=SEED)

    # Define Pipe
    pipeline= Pipeline(steps = [('preprocess', preprocessor),
                                ('pca', PCA()),
                                ('reg', reg)])
    

    # Define RandomizedSearch
    model = GridSearchCV(estimator=pipeline,
                         param_grid = grid_params,
                         cv = kf,
                         scoring='r2',
                         refit=True,
                         n_jobs=-1,
                         verbose=2) 
    model.fit(X,y)
    
    # evaluate
    print('Best Parameters:', model.best_params_)
    print('\nBest Score:', model.best_score_)
    
    return reg,model.best_score_
```

```python
# Define parameters
param_dicts = [
    {'reg': RandomForestRegressor(),
        'name': 'RF',
        'grid_dict': {
            'pca__n_components': [6,7,8,9],
            'reg__criterion': ['absolute_error', 'squared_error'],
            'reg__max_depth': [3,5],
            'reg__min_samples_leaf': [2,3],
            'reg__min_samples_split': [5,7,10],
            'reg__n_estimators': [100],
            'reg__random_state': [SEED]}
    },

    {'reg': XGBRegressor(),
        'name': 'XGB',
        'grid_dict': {
            'pca__n_components': [6,7,8,9],
            'reg__n_estimators': [100, 200],
            'reg__max_depth': [5,10],
            'reg__learning_rate': [0.01,0.03],
            'reg__gamma': [1,3,5],
            'reg__random_state': [SEED]}
    },
    
    {'reg': KNeighborsRegressor(),
      'name': 'KNN',
       'grid_dict': {
           'pca__n_components': [6,7,8,9],
           'reg__n_neighbors':[3,5,8,9,10]}
      }
        
    ]
```

```python
# Loop through each model and find best parameters
best_models=[]

for el in tqdm(param_dicts):
    print(el['name'])
    model_name, best_score = model_loop(X_train, y_transformed_train, el['reg'], el['grid_dict'])
    best_models.append([model_name,best_score])
```
