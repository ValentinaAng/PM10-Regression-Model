#### GridSearchCV

```python
def model_loop(X, y, reg, grid_params):
    
    print('Currently fitting model:', reg)
    
    # Define KF
    kf = KFold(n_splits=3,shuffle=True, random_state=SEED)

    # Define Pipe
    pipeline= Pipeline(steps = [('preprocess', preprocessor),
                                ('select_kbest', SelectKBest(f_regression)),
                                ('reg', reg)])
    

    # Define RandomizedSearch
    model = GridSearchCV(estimator=pipeline,
                         param_grid = grid_params,
                         cv = kf,
                         scoring='r2',
                         refit=True,
                         n_jobs=-1,
                         verbose=2) 
    model.fit(X,y)
    
    # evaluate
    print('Best Parameters:', model.best_params_)
    print('\nBest Score:', model.best_score_)
    
    return reg,model.best_score_
```

```python
# Define parameters
param_dicts = [
    {'reg': RandomForestRegressor(),
        'name': 'RF',
        'grid_dict': {
            'select_kbest__k': [12,14],
            'reg__criterion': ['absolute_error', 'squared_error'],
            'reg__max_depth': [5,6],
            'reg__min_samples_leaf': [2,3],
            'reg__min_samples_split': [3,5],
            'reg__n_estimators': [100,150],
            'reg__random_state': [SEED]}
    },

    {'reg': XGBRegressor(),
        'name': 'XGB',
        'grid_dict': {
            'select_kbest__k': [10,12,14],
            'reg__n_estimators': [100, 200],
            'reg__max_depth': [7,10],
            'reg__learning_rate': [0.05],
            'reg__gamma': [4,5],
            'reg__reg_lambda':[2,3,5],
            'reg__random_state': [SEED]}
    },
    
    {'reg': KNeighborsRegressor(),
      'name': 'KNN',
       'grid_dict': {
           'select_kbest__k': [10,12,14],
           'reg__n_neighbors':[5,10,15]}
      }
        
    ]
```

```python
# Loop through each model and find best parameters
best_models=[]

for el in tqdm(param_dicts):
    print(el['name'])
    model_name, best_score = model_loop(X_train, y_transformed_train, el['reg'], el['grid_dict'])
    best_models.append([model_name,best_score])
```
